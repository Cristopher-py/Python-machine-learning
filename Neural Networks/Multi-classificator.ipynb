```python
# Importing libraries
import torch

from sklearn.datasets import make_blobs
from sklearn.model_selection import train_test_split

import matplotlib.pyplot as plt
```

```python
# Cluster dataset
xy, z = make_blobs(n_samples = 1000,
                   n_features = 2,
                   centers = 4,
                   cluster_std = 1.5,
                   center_box = (- 10.0, 10.0),
                   shuffle = True,
                   random_state = 42,
                   return_centers = False)
```

```python
# Visualizing it
plt.figure(figsize = (8, 6))
plt.scatter(xy[:, 0], xy[:, 1], c = z, cmap = 'RdYlBu', alpha = 0.75, edgecolors = 'black')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Dataset (4 classes)')
plt.colorbar(label = 'Class')
plt.grid(True, alpha = 0.5)
plt.show()
```

```python
# Conversion to torch tensors
xy = torch.from_numpy(xy).type(torch.float)
z = torch.from_numpy(z).type(torch.float)
```

```python
# Splitting
xy_train, xy_test, z_train, z_test = train_test_split(xy, z,
                                                      test_size = 0.2,
                                                      random_state = 42,
                                                      shuffle = True)
```

```python
xy_test
```

```python
z_test
```

```python
# Neural Network model
model = torch.nn.Sequential(
    
    torch.nn.Linear(in_features = 2,
                    out_features = 8),
                    
    torch.nn.ReLU(),
                    
    torch.nn.Linear(in_features = 8,
                    out_features = 8),
                                    
    torch.nn.ReLU(),
    
    torch.nn.Linear(in_features = 8,
                    out_features = 4)
)
```

```python
# Loss function
loss_fn = torch.nn.CrossEntropyLoss()
```

```python
# Optimizer
optimizer = torch.optim.SGD(params = model.parameters(),
                            lr = 0.01)
```

```python
# Random predictions
torch.softmax(model(xy_train), dim = 1).argmax(dim = 1)
```

```python
# Training phase
z_train = z_train.long()
z_test = z_test.long()

train_losses = []
test_losses = []
epochs = 2_000

for epoch in range(epochs):

    model.train()
    z_logits = model(xy_train)
    
    train_loss = loss_fn(z_logits, z_train)
    optimizer.zero_grad()
    train_loss.backward()
    optimizer.step()
    
    model.eval()
    with torch.inference_mode():
        z_test_logits = model(xy_test)
        test_loss = loss_fn(z_test_logits, z_test)
    
    train_losses.append(train_loss.item())
    test_losses.append(test_loss.item())
    
    # Print progress every 100 epochs
    if epoch % 50 == 0:
        print(f"Epoch: {epoch:4d} | Train Loss: {train_loss:.5f} | Test Loss: {test_loss:.5f}")

# Plot loss curves
plt.figure(figsize = (10, 5))
plt.plot(train_losses, label='Train Loss', alpha = 0.8)
plt.plot(test_losses, label='Test Loss', alpha = 0.8)
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training and Test Loss Over Time')
plt.legend()
plt.grid(True, alpha = 0.3)
plt.show()
```

```python
# Accuracy test, test points
torch.softmax(model(xy_test), dim = 1).argmax(dim = 1) == z_test
```

```python
# Accuracy test, train points
torch.softmax(model(xy_train), dim = 1).argmax(dim = 1) == z_train
```
